### Sequences

These are built in objects like list,tuples, strings etc

```python
my_sequence = "Senior Engineier"
print(len(my_sequence))
print (my_sequence[-15])
```

### **index()**

```python
output = my_sequence.index('S')
print(output)
output = my_sequence.index(' ',-15, -1 )
print(output)
```

### **[*start*, *stop*, *step*]** _Note: stop index is not included_

**Start**: is the starting index and it is included in the output  
__Stop__: is the end index but not included  
**Step**: how many index from the start should be include. if your start is at index 0 and step is 4, then included index will be 0+4, 4+4, and 8+4


```python
substring = my_sequence[0:6]
print(substring)
substring = my_sequence[0:13:4]
print(substring)
```

### **Min, Max, Count** 

Works only on int/double/float values in a sequence or list

```python
value = [0, 1, 2, 4, 5, 6]
print(min(value))
print(value.count(5))
print(max(value))
```

```python
def test_list_comprehension(list_values):
    square_root = [value * value for value in list_values if value >= 0]
    return square_root
    
```

```python
new_list = [test_list_comprehension, 'andy', True, False, 1, []]
```
